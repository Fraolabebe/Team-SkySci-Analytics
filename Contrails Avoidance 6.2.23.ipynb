```python
from datetime import datetime, timedelta
month = 1
day = 1

# Define the number of days in each month
days_in_month = {
    1: 31,  # January
    2: 28,  # February
    3: 31,  # March
    4: 30,  # April
    5: 31,  # May
    6: 30,  # June
    7: 31, # July
    8: 31, #August
    9: 30, #September
    10: 31, #October
    11: 30, #Novemeber
    12: 31 #December
    # Add the remaining months...
}

# Iterate through the months and days
for _ in range(365):  # Assuming a non-leap year
    # Print the current month and day
    
    print(f"Month: {month}, Day: {day}")
    # Define the API endpoint URL
    url = 'https://www.wunderground.com/history/daily/us/va/sterling/KIAD/date/2022-' + str(month) + "-" + str(day)
    # Make a GET request to the API endpoint
    response = requests.get(url)
    print(response.text)
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find the container that holds the line chart
    chart_container = soup.find('div', class_='charts-canvas')
    print(chart_container)
    

    # Find the line chart for temperature data
    temperature_chart = chart_container.find('div', class_='plot.temperature.line')
    
    print(temperature_chart)

    # Increment the day
    if day < days_in_month[month]:
        day += 1
    else:
        # If the current day exceeds the number of days in the month, reset the day to 1
        day = 1

        # Increment the month
        if month < 12:
            month += 1
        else:
            # If the current month is December, reset the month to January
            month = 1
```
