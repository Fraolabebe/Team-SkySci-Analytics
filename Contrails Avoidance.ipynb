# Web SCraping


```python
# Define the API endpoint URL
url = 'https://www.wunderground.com/history/daily/us/va/sterling/KIAD/date/2022-1-1'
# Make a GET request to the API endpoint
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python

# Find the container that holds the line chart
chart_container = soup.find('div', class_='chart-container')

# Find the line chart for temperature data
temperature_chart = chart_container.find('div', class_='plot.temperature.line')

# Extract the hourly temperature data points
temperature_data = temperature_chart['d']
temperature_values = temperature_data.split(',')

# Print the hourly temperature values
for hour, value in enumerate(temperature_values):
    print(f"Hour {hour + 1}: {value}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the date selector container
date_selector = soup.find('div', class_='date-selector ng-star-inserted')
date = date_selector.find('input')['value']

# Find the line chart container
chart_container = soup.find('div', class_='wc-base-chart plot')
path_data = chart_container.find('path')['d']

# Split the path data into individual points
points = path_data.split('L')

# Extract the x-axis grid labels
hours = []
for point in points:
    _, y = point.split(',')
    if float(y) == 120.0:  # Assuming y-coordinate of 120.0 indicates a new hour
        hour = point.split(',')[0]
        hours.append(hour)

# Create a dataframe with the extracted data
df = pd.DataFrame({'Hour': hours})
df['Date'] = date

# Print the dataframe
print(df)
```
