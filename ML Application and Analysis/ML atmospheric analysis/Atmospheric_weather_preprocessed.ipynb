```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"C:\Users\abebe\Documents\George Mason University\6- SUMME 2023\Team-SkySci-Analytics\Atmospheric Data Scraper\clean_atmospheric_data.csv"
```

```python
data = pd.read_csv(path)
data
```

```python
# df.drop('Timestamp', axis=1, inplace=True)
df= data[["Temperature", "Relative Humidity", 'Level', "Image Present"]]
df = df.rename(columns={"Temperature": "temperature", "Relative Humidity":"humidity", 'Level':'pressure', "Image Present": "image"})
df.head()
```

```python
#Normalized Data
df['temperature'] = df['temperature'] / df['temperature'].max()
df['humidity'] = df['humidity'] / df['humidity'].max()
df['pressure'] = df['pressure'] / df['pressure'].max()
df.head()
```

```python
df.to_csv("atmospheric_weather_data_preprocessed.csv", index=False)
```
